### Resampling

### Split dataset

```python
# Split train and test
X_train, X_test, y_train, y_test = train_test_split(X,y)
X_train_smote, X_test_smote, y_train_smote, y_test_smote = train_test_split(X_SMOTE_resampled, y_SMOTE_resampled)
X_train_adasyn, X_test_adasyn, y_train_adasyn, y_test_adasyn = train_test_split(X_ADASYN_resampled, y_ADASYN_resampled)
```

```python
# Standarisasi
scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

X_train_smote = scaler.fit_transform(X_train_smote)
X_test_smote = scaler.transform(X_test_smote)

X_train_adasyn = scaler.fit_transform(X_train_adasyn)
X_test_adasyn = scaler.transform(X_test_adasyn)
```

# Training model

```python
def train_model(model, X_training, y_training, X_testing, modelName, y_testing, methodName):
    model.fit(X_training,y_training)
    y_pred = model.predict(X_testing)

    # Evaluate model
    print(f"{modelName} Model")
    acc = round(accuracy_score(y_testing, y_pred), 3)
    print("Accuracy:", acc)
    print("Classification Report:")
    print(classification_report(y_testing, y_pred))

    #Confussion_matrix
    cm = confusion_matrix(y_testing, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
    plt.title(f'Confusion Matrix {modelName}')
    plt.xlabel('True')
    plt.ylabel('Predict')
    plt.show()

    print("\n")
    if (acc > 0.9):
    # Save the model using pickle
        with open(f"{modelName}_model_{methodName}.pkl", "wb") as model_file:
            pickle.dump(model, model_file)
```

```python
knn_model = KNeighborsClassifier(n_neighbors=3)
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
xgb_model = XGBClassifier(learning_rate=0.1, n_estimators=100, random_state=42)
```

```python
## KNN ##
nameModel = "K-Nearest Neighbors (KNN)"
nonRes_method = "Non Resampling"
smote_method = "SMOTE"
adasyn_method = "ADASYN"
# Model tidak resampling dan tidak normalisasi
train_model(knn_model, X_train, y_train, X_test, nameModel, y_test, nonRes_method)
# Model SMOTE
train_model(knn_model, X_train_smote, y_train_smote, X_test_smote, nameModel, y_test_smote, smote_method)
# Model ADASYN
train_model(knn_model, X_train_adasyn, y_train_adasyn, X_test_adasyn, nameModel, y_test_adasyn, adasyn_method)
```

```python
## Random Forest ##
nameModel = "Random Forest"
nonRes_method = "Non Resampling"
smote_method = "SMOTE"
adasyn_method = "ADASYN"
# Model tidak resampling dan tidak normalisasi
train_model(rf_model, X_train, y_train, X_test, nameModel, y_test, nonRes_method)
# Model SMOTE
train_model(rf_model, X_train_smote, y_train_smote, X_test_smote, nameModel, y_test_smote, smote_method)
# Model ADASYN
train_model(rf_model, X_train_adasyn, y_train_adasyn, X_test_adasyn, nameModel, y_test_adasyn, adasyn_method)
```

```python
## Random Forest ##
nameModel = "XGBoost"
nonRes_method = "Non Resampling"
smote_method = "SMOTE"
adasyn_method = "ADASYN"
# Model tidak resampling dan tidak normalisasi
train_model(xgb_model, X_train, y_train, X_test, nameModel, y_test, nonRes_method)
# Model SMOTE
train_model(xgb_model, X_train_smote, y_train_smote, X_test_smote, nameModel, y_test_smote, smote_method)
# Model ADASYN
train_model(xgb_model, X_train_adasyn, y_train_adasyn, X_test_adasyn, nameModel, y_test_adasyn, adasyn_method)
```
